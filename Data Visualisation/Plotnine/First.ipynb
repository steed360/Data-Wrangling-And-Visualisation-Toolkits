```python
from plotnine import ggplot, geom_point, aes, stat_smooth, facet_wrap
from plotnine.data import mtcars

fig = plt.figure(figsize=(40,7), facecolor = "black")

gg = (ggplot(mtcars, aes('wt', 'mpg', color='factor(gear)'))
 + geom_point()
 + stat_smooth(method='lm')
 + facet_wrap('~gear'))

fig = plt.figure(figsize=(40,7), facecolor = "black")

fig = gg.draw()
axes = fig.get_axes()

for ax in axes:
    #Axes (ax).spline['top']
    #xaxis = ax.get_xaxis()
    print (1)

help (ax)
    

    
 
```
